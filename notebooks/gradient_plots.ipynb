```python
data = np.load('../Synth_Datasets/range100.npy')
# data = np.load('../Synth_Datasets/skewed.npy')
data = np.arange(-50, 50)
data = torch.from_numpy(data.astype(float))
```

```python
y_hat_arr = np.linspace(data.min(), data.max(), len(data))
q = 0.7
```

```python
def collect_loss_grad(loss_fn, y_hat_arr):
    loss_list = []
    grad_list = []
    for y_hat in y_hat_arr:
        curr_pred = Variable(torch.tensor([y_hat]), requires_grad=True)
        if curr_pred.grad is not None:
            curr_pred.grad = 0.0
        loss = loss_fn(data, curr_pred, q=q)
        loss.backward()
        loss_list.append(loss.item())
        grad_list.append(-1*curr_pred.grad.item())
        del curr_pred
    return loss_list, grad_list
```

```python
def get_zero_idx(grad_list):
    zero_idx = []
    for idx in range(len(grad_list)-1):
        if (grad_list[idx] < 0) and (grad_list[idx+1] >= 0):
            zero_idx.append(idx)
        if (grad_list[idx] >= 0) and (grad_list[idx+1] < 0):
            zero_idx.append(idx)
    return zero_idx
```

```python
def get_est_grads(loss_list, y_hat_arr):
#     import pdb; pdb.set_trace()
    grad_ests = []
    for idx in range(len(loss_list)-1):
        curr_est = ((loss_list[idx+1] - loss_list[idx])/
                    (y_hat_arr[idx+1] - y_hat_arr[idx]))
        grad_ests.append(-1*curr_est)
    return grad_ests
```

```python
def get_coverages(y, y_hat_arr):
    cov_list = []
    for y_hat in y_hat_arr:
        cov = torch.mean(y <= y_hat, dtype=float).item()
        cov_list.append(cov)
    return cov_list
```

```python
def plot_loss_grads(title, y, y_hat_arr, grads, losses, 
                    zero_grads, grad_ests=None):
    
    fig, ax = plt.subplots()
    
    ax.plot(y_hat_arr, grads, c='r', label='grads')
    ax.axhline(0, c='r')
    arg_zero_grads = []
    for zero_idx in zero_grads:
        ax.scatter(y_hat_arr[zero_idx], 0, marker='*', s=200, c='r', label='zero grad')
        arg_zero_grads.append(round(y_hat_arr[zero_idx], 2))
        
    if grad_ests is not None:
        ax.plot(y_hat_arr[:-1], grad_ests, '--', label='grad_ests', c='y')
    ax_range_eps = (np.max(grads) - np.min(grads))/20.0
    try:
        ax.set_ylim(np.min(grads) - ax_range_eps, np.max(grads) + ax_range_eps)
    except:
        pass
    
    ax1 = ax.twinx()
    ax1.plot(y_hat_arr, losses, c='b', label='losses')
    ax1.scatter(y_hat_arr[np.argmin(losses)], np.min(losses), 
                c='b', label='loss argmin')
    ax1_range_eps = (np.max(losses) - np.min(losses))/20.0
    try:
        ax1.set_ylim(np.min(losses)-ax1_range_eps, np.max(losses)+ax1_range_eps)
    except:
        pass
    cov_zero_grad = np.mean(y.numpy() <= y_hat_arr[zero_grads[0]])
    cov_min_loss = np.mean(y.numpy() <= y_hat_arr[np.argmin(losses)])
    
    ax.set_title('{}: {:.2f} below zero grad, {:.2f} below min loss'.format(title, cov_zero_grad, cov_min_loss))
    ax.set_xlabel('y_hat')
    ax.set_ylabel('gradient')
    ax1.set_ylabel('loss')
    ax.legend(loc=1)
    ax1.legend(loc=2)
    
    plt.show()
```

```python
def do_all(title, loss_fn, y_hat_arr, y):
    losses, grads = collect_loss_grad(loss_fn, y_hat_arr)
    zero_grads = get_zero_idx(grads)
    grad_ests = get_est_grads(losses, y_hat_arr)

    plot_loss_grads(title, y, y_hat_arr, 
                    grads, losses, zero_grads, grad_ests)
```
