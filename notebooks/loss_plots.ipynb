```python
data = np.load('../Synth_Datasets/range100.npy')
# data = np.load('../Synth_Datasets/skewed.npy')
```

```python
q = 0.7
```

```python
pts_under_list = []
pts_over_list = []
diff_under_list = []
diff_over_list = []
diff_all_list = []
cov_list = []
cov_diff_list = []

y_hat_arr = np.linspace(data.min(), data.max(), 150)

for y_hat in y_hat_arr:
    idx_under = (data <= y_hat)
    idx_over = ~idx_under
    pts_under = data[idx_under]
    pts_over = data[idx_over]
    
#     import pdb; pdb.set_trace()
    diff_under = np.mean(y_hat - pts_under) if pts_under.size > 0 else 0
    diff_over = np.mean(y_hat - pts_over) if pts_over.size > 0 else 0
    diff_all = np.sum(data - y_hat)
    cov = np.mean(idx_under)
    cov_diff = cov - q
    
    pts_under_list.append(pts_under)
    pts_over_list.append(pts_over)
    diff_under_list.append(diff_under)
    diff_over_list.append(diff_over)
    diff_all_list.append(diff_all)
    cov_list.append(cov)
    cov_diff_list.append(cov_diff)

# diff_under_arr = np.asarray(diff_under_list)
# diff_over_arr = np.asarray(diff_over_list)
# diff_all_arr = np.asarray(diff_all_list)
# cov_arr = np.asarray(cov_list)
# cov_diff_arr = np.asarray(cov_diff_list)
```
